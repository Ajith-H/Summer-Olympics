```python
import pandas as pd
olympics = pd.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
print("Number of cities : ",len(olympics['City'].unique()))

# to print each unique city occurring in the city column
# for cities in olympics['City'].unique(): print(cities)
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
olympics[olympics['Medal'] == 'Gold'].groupby('Sport').size().sort_values(ascending=False).head().plot.bar()
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
olympics.groupby('Sport').size().sort_values(ascending=False).head().plot.bar()
```

### 4. Which player has won most number of medals? (Top 5)

```python
olympics.groupby('Athlete').size().sort_values(ascending=False).head().plot.bar()
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
olympics[olympics['Medal'] == 'Gold'].groupby('Athlete').size().sort_values(ascending=False).head().plot.bar()
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
olympics[['Year']][(olympics['Country'] == 'IND') & (olympics['Medal'] == 'Gold')].min()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
olympics.groupby('Event').size().sort_values(ascending=False).head().plot.bar()
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
olympics[olympics['Gender']=='Women'].groupby('Sport').size().sort_values(ascending=False).head().plot.bar()
```
